```
from flask import Flask,render_template,request
import requests
from predict import summarize
from flask_ngrok import run_with_ngrok

app=Flask(__name__)
run_with_ngrok(app)
# api-endpoint
@app.route("/")
def home():
    URL = "http://newsapi.org/v2/top-headlines?country=in&apiKey=2e0ec8c57e8e4da7aef1822653d387ec"
    r = requests.get(url = URL)
    data = r.json()
    #title=data["articles"]["title"]
    #discriptions=data["articles"]["description"]
    return render_template("Home.html",data=data)

@app.route("/text",methods=["POST","GET"])
def predict():
    return render_template("text.html")

@app.route("/output",methods=["POST","GET"])
def output():
    text =request.form.get("text")
    output=summarize(paragraph=text)
    return render_template("text.html", output=output)

app.run()
```
